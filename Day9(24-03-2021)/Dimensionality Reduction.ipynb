### Dimensionality Reduction
- Dimensions are represented as features or columns in dataset
- Reduction of no of features in our dataset
    1. Feature Elimination: Removing unimportant features or columns.
    2. Feature Extraction: We can create a new columns where each new feature is combination of old columns
        - Eg: Feature names--> Product_price, No_of_Products, Cost_products
        - Final_price = Product_price *  No_of_Products * Cost_products
        
        
- **Advantage**
- By combining the features you can not loos information by reducing the dimensions

### Principal Component Analysis(PCA)
- PCA is one of the technique for the feature extarction in Dimentionality reduction.

### Apply Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_transformed = scaler.fit_transform(X)
X_transformed
```

```python
pd.DataFrame(X_transformed)
```

```python
pd.DataFrame(X_transformed).describe()
```

### Apply PCA model

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
newFeature = pca.fit_transform(X_transformed)
```

```python
pca_df = pd.DataFrame(newFeature,columns = ['PCA1','PCA2'])
pca_df.head()
```

```python
pca.explained_variance_ratio_
```
