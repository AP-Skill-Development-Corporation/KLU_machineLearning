### Clustering
- Clusterning can be defined as the task of identifying the subgroups in the given dataset.

### K Means Clustering Model
- K Means is simple and easy way to classify the given dataset through a certrain no of clusters.
- K means algorithm takes a dataset with constant value k and returns k centroids.
    - Centroid defines the cluster of data in the dataset which are similar to one another.

### Working on K Means
- Step1: Randomly take K no of clusters(data points)
- Step2: Calculate the Distance
- Step3: Calculate new Clusters
- Step4: Divide into groups
- Step5: Repeat same procedure

### Distance formula
- Euclidian distance between two points is
    - sqrt( (x2-x1)^2 + (y2-y1)^2 )

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/IRIS.csv')
df.head()
```

```python
X = df.iloc[:, [0,1,2,3]].values
X
```

```python
from sklearn.cluster import KMeans
kmeans5 = KMeans(n_clusters = 5)
```

```python
y_kmeans5 = kmeans5.fit_predict(X)
```

```python
y_kmeans5
```

### Elbow Method

```python
Error = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i).fit(X)
    kmeans.fit(X)
    Error.append(kmeans.inertia_)
plt.plot(range(1,11), Error)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans3 = KMeans(n_clusters = 3)
y_kmeans3 = kmeans3.fit_predict(X)
print(y_kmeans3)
```
